## 2. Load and Explore the Dataset
We load the AG News dataset which contains two text columns (Title and Description) 
and a Class Index label with four categories:
- 1 = World
- 2 = Sports
- 3 = Business
- 4 = Science/Technology

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

class_names = {1: 'World', 2: 'Sports', 3: 'Business', 4: 'Science/Tech'}
train_df['Category'] = train_df['Class Index'].map(class_names)
test_df['Category'] = test_df['Class Index'].map(class_names)

print("Training set shape:", train_df.shape)
print("Test set shape:", test_df.shape)
print("\nClass Distribution (Training):")
print(train_df['Category'].value_counts())
print("\nSample rows:")
train_df.head(3)
```
